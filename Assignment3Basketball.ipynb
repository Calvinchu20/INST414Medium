```python
import pandas  as pd 
import csv
from scipy.sparse import lil_matrix 
import scipy.spatial.distance
from sklearn.metrics import DistanceMetric
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
```

```python
# creating the dataFrame from the Csv file 
df = pd.read_csv('cbb23.csv')
df = df.set_index('TEAM')
df = df.drop(columns=['CONF','WAB','POSTSEASON','SEED'])
df.fillna(0)
```

```python
# establishing the query for similarity uses 
query1 = ['ADJOE','ADJDE','BARTHAG']
query2 = ['TOR','ORB']
query3 = ['EFG_O','2P_O','3P_O']
```

```python
# establishing the target value and creating subdataframe based on the queries
target =  'Connecticut'
df1 = df.loc[:, query1]
df2 = df.loc[:, query2]
df3 = df.loc[:, query3]
df1.values.reshape(1,-1)
df2.values.reshape(1,-1)
df3.values.reshape(1,-1)
```

```python
# query 1 similary metric
target_query1 =  df1.loc[target] 
tarq1 =  dict()
for i, row in df1.iterrows():
    rave = cosine_similarity([row],[target_query1])
    tarq1[i] = float(rave)
  
tarq1 =  sorted(tarq1.items(), key = lambda x:x[1],reverse=True)
count = 0
print('top 10 similar college basketball team on  to Connecticut using Cosine similarity')
for x in tarq1:
    if count == 10:
        break
    else:
        if x[0] == 'Connecticut':
            continue
        else:
            print(x[0],' ',x[1])
            count+=1

```

```python
# query 2 similary metric
target_query2 = df2.loc[target]
tarq2 = dict()
for i, row in df2.iterrows():
    rave = cosine_similarity([row],[target_query2])
    tarq2[i] = float(rave)
  
tarq2 =  sorted(tarq2.items(), key = lambda x:x[1],reverse=True)
count = 0
print('top 10 similar college basketball team on rebound and turnover rates to Connecticut using Cosine similarity')

for x in tarq2:
    if count == 10:
        break
    else:
        if x[0] == 'Connecticut':
            continue
        else:
            print(x[0],' ',x[1])
            count+=1
```

```python
# query 3 similary metric
target_query3 = df3.loc[target]
tarq3 = dict()
for i, row in df3.iterrows():
    rave = euclidean_distances([row],[target_query3])
    tarq3[i] = float(rave)
  
tarq3 =  sorted(tarq3.items(), key = lambda x:x[1],reverse=False)
count = 0
print('top 10 similar college basketball team based on scoring(field goals, 2 point percentage, 3 point percentage) to Connecticut using  euclidean distance ')
for x in tarq3:
    if count == 10:
        break
    else:
        if x[0] == 'Connecticut':
            continue
        else:
            print(x[0],' ',x[1])
            count+=1
```
